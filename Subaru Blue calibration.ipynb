# Calculate photometry

```python
phot_table_Hubble2 = aperture_photometry(image_data_w475_fireball, apetures_phot_hubble2)
phot_table_Hubble = aperture_photometry(image_data_w475_fireball, apetures_phot_hubble)
```

```python
Hubble_counts = phot_table_Hubble["aperture_sum"]/apetures_phot_hubble.area
Hubble_counts2 = phot_table_Hubble2["aperture_sum"]/apetures_phot_hubble2.area
```

```python
phot_table_subaru = aperture_photometry(image_data_subaru_blue, phot_apeture_subaru)
phot_table_subaru2 = aperture_photometry(image_data_subaru_blue, phot_apeture_subaru2)
```

```python
Subaru_counts = phot_table_subaru["aperture_sum"]/phot_apeture_subaru.area
Subaru_counts2 = phot_table_subaru2["aperture_sum"]/phot_apeture_subaru2.area
```

```python
plt.scatter(Hubble_counts, Subaru_counts, label = "3 arcsecond radius")
plt.scatter(Hubble_counts2, Subaru_counts2, label = "5 arcsecond radius")

best_fit = np.poly1d(np.polyfit(Hubble_counts, Subaru_counts, 1))
plt.plot(Hubble_counts, best_fit(Hubble_counts), label = "best fit line of 3 arcsecond radius")

best_fit2 = np.poly1d(np.polyfit(Hubble_counts2, Subaru_counts2, 1))
plt.plot(Hubble_counts2, best_fit2(Hubble_counts2), label = "best fit line of 5 arcsecond radius")

from astropy.table import Table, vstack
Hubble_sums = vstack([Hubble_counts, Hubble_counts2])
subaru_sums = vstack([Subaru_counts, Subaru_counts2])

best_fit3 = np.poly1d(np.polyfit(Hubble_sums["aperture_sum"] , subaru_sums["aperture_sum"], 1))
plt.plot(Hubble_sums["aperture_sum"], best_fit3(Hubble_sums["aperture_sum"]), label = "best fit line overall")

plt.legend()
plt.xlabel("Hubble photometry counts")
plt.ylabel("Subaru photometry counts")
```

```python
plt.scatter(Hubble_counts2, Subaru_counts2, label = "5 arcsecond radius")
best_fit2 = np.poly1d(np.polyfit(Hubble_counts2, Subaru_counts2, 1))
plt.plot(Hubble_counts2, best_fit2(Hubble_counts2), label = "best fit line of 5 arcsecond radius")

plt.legend()
plt.xlabel("Hubble photometry counts")
plt.ylabel("Subaru photometry counts")
```
